```python
import gpt_2_simple as gpt2

model_name = "124M"
gpt2.download_gpt2(model_name=model_name)   # model is saved into current directory under /models/124M/

sess = gpt2.start_tf_sess()
gpt2.finetune(sess,
              dataset='song_lyrics2.csv',
              model_name=model_name,
              run_name='run_lyrics',
              steps=1000,
              save_every=200,
              sample_every=25)

gpt2.generate(sess)
```
